```python
import plotly.express as px

df = px.data.gapminder().query("country=='Thailand'")

fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')
fig.show()
```

```python
df = px.data.gapminder().query("continent=='Asia'")
fig = px.line(df, x="year", y="lifeExp", color='country')
fig.show()
```

```python
import plotly.express as px
df = px.data.stocks(indexed=True)
#print(df.head())
fig = px.line(df, facet_col="company", facet_row_spacing=0.01, height=200, width=800)

# hide and lock down axes
fig.update_xaxes(visible=False, fixedrange=True)
fig.update_yaxes(visible=False, fixedrange=True)

# remove facet/subplot labels
#fig.update_layout(annotations=[], overwrite=True)

# strip down the rest of the plot
fig.update_layout(
    showlegend=False,
    plot_bgcolor="aliceblue",
    margin=dict(t=20,l=10,b=10,r=10),
    paper_bgcolor="pink",
    font=dict(size=18, color="black", family="Times New Roman")
)

# disable the modebar for such a small plot
fig.show(config=dict(displayModeBar=False))
```

```python
import plotly.graph_objects as go
import numpy as np

x = np.arange(10)
print(x)

fig = go.Figure(data=go.Scatter(x=x, y=x**2))
fig.show()
```

```python
import plotly.graph_objects as go

# Create random data with numpy
import numpy as np
np.random.seed(1)

N = 100
random_x = np.linspace(0, 1, N)
#print(random_x)
random_y0 = np.random.randn(N)
#print(random_y0)
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N)

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=random_x, y=random_y0,
                    mode='lines',
                    name='lines'))
fig.add_trace(go.Scatter(x=random_x, y=random_y1,
                    mode='lines+markers',
                    name='lines+markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y2,
                    mode='markers', name='markers'))

fig.show()
```
